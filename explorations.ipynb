```python
import numpy as np
import pandas as pd
from hebo.design_space.design_space import DesignSpace
```

```python
# Define the design space with a continuous variable x0 and an integer variable x1.
space = DesignSpace().parse([
    {'name': 'x0', 'type': 'num', 'lb': -3, 'ub': 3},
    {'name': 'x1', 'type': 'int', 'lb': 0, 'ub': 5},
])

def evaluate(xs: pd.DataFrame):
    x0 = xs["x0"]
    x1 = xs["x1"]
    # Example oracle combining quadratic, sinusoidal, and linear terms
    ys = x0**2 + 0.5 * np.sin(2 * np.pi * (x1 - 0.2)) + 0.25 * x0
    return ys.values
```
